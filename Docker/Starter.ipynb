## Stopping Containers

We can do this from Two Ways



These both stops the containers. 

**Whats the difference in here?? **

## When we issue docker stop command a hardware signal is sent to the process. In the case of *docker stop* it sends a sigterm message which is short for terminate signal (message recvied by process telling essentially to shut down on its own time. And provide little bit time for the cleanup to the container Ex saving file,sending results etc)




## Docker Kill will not give time for cleanup and stuff . It terminates everything immediatly



## Pro Tip: If after docker stop our container won't stop with in 10 sec then it will automatically fall to docker kill
