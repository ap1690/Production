## Any time that we talk about an image we're really talking about the file system snapshot(kind of like a copy paste of a very specific set of directories or files)



## So we might have image that contains just chrome and python . An image will specifically contains a specific startup command . Image is just a file system snapshot with a startup command

Let's see what happens when we try to create container from an image



## First off the kernal is going to isolate a little section of the harddrive and make it available to just the container which is on built . After that a little subset is created from the file snapshot inside the image and then the startup command is executed which will start up the process
